### Data Visualization with Matplotlib:---

**Matplotlib** is a powerful plotting library in python that enables the creation of static, animated and interactive visualisations. It widely used for data visualisation in data science and analytics . 

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]
z = [1,8,27,64,125]
```

```python
plt.title('Square of Numbers')

plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.plot(x,y,color='red',linewidth=3,marker='*',label='Square')
plt.plot(x,z,color='blue',linewidth=3,marker='o',linestyle='--',label='Cube')
plt.grid()

plt.legend()  # <-- this makes labels visible
plt.show()
```

#### Multiple plots:

#### Sample data:-

```python
x = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [1,8,27,64,125]
```

```python
plt.figure(figsize=(9,5))

plt.subplot(1,2,1)
plt.title('Plot 1')
plt.plot(x,y1,color='green',linewidth=3,label='Plot 1')
plt.legend()

plt.subplot(1,2,2)
plt.title('Plot 2')
plt.plot(x,y2,color='blue',linewidth=3,label='Plot 2')
plt.legend()

plt.show()
```

#### Another plotting example:--

```python
plt.figure(figsize=(9,5))

plt.subplot(2,2,1)
plt.title('Plot 1')
plt.plot(x,y1,color='green',linewidth=3,label='Plot 1')
plt.legend()

plt.subplot(2,2,2)
plt.title('Plot 2')
plt.plot(x,y2,color='blue',linewidth=3,label='Plot 2')
plt.legend()

plt.subplot(2,2,3)
plt.title('Plot 3')
plt.plot(y1,x,color='red',linewidth=3,label='Plot 3')
plt.legend()

plt.subplot(2,2,4)
plt.title('Plot 4')
plt.plot(y2,x,color='pink',linewidth=3,label='Plot 4')
plt.legend()

plt.show()
```

### Bar Plot:-

```python
categories = ['A','B','C','D','E']
values = [5,8,4,9,7]

plt.bar(categories,values,color='green',width=0.5)
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

### Histograms:--

**Histograms** are used to represent the distribution of a dataset. They divide the data into bins and count the number of data points in each bin.

```python
data = [1,2,2,3,3,3,4,4,5,5,5,5,5]

plt.hist(data,bins=5,color='green',edgecolor='black')
# plt.show()
```

### Scatter plot:--

```python
# sample values:--
x = [1,2,3,4,5]
x = [2,3,4,5,6]

plt.scatter(x,y,color='green',marker = '*')
```

#### Pie Charts:--

```python
labels = ['A','B','C','D','E']
sizes = [10,30,20,15,25]
colors = ['green','red','yellow','pink','blue']
explode = (0.2,0,0,0,0)

plt.pie(sizes,labels=labels,colors=colors,autopct='%1.2f%%',explode=explode)
```

### Sales Data Visualization:--

```python
import pandas as pd
sales_data_df = pd.read_csv('sales_data.csv')
sales_data_df.head()
```

```python
sales_data_df.info()
```

#### Total sales by Product:--

```python
total_sales_by_product = sales_data_df.groupby('Product Category')['Total Revenue'].sum()
total_sales_by_product
```

```python
total_sales_by_product.plot(kind='bar',color='green')
```

#### Plot trend over time:--

```python
sales_trend = sales_data_df.groupby('Date')['Total Revenue'].sum().reset_index()   #reste_index will give separate columns.
sales_trend
```

```python
plt.plot(sales_trend['Date'],sales_trend['Total Revenue'])
```
